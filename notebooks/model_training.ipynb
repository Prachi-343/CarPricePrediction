```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("./data/carprice.csv", encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data=data.drop(labels=["customer e-mail","country","gender","age","customer name"],axis=1)
```

```python
data.head()
```

```python
x=data.drop(labels="car purchase amount",axis=1)
```

```python
x
```

```python
y=data[["car purchase amount"]]
```

```python
y
```

```python
numerical_cols=x.select_dtypes(exclude="object").columns
```

```python
numerical_cols
```

```python
from sklearn.impute import SimpleImputer ## HAndling Missing Values
from sklearn.preprocessing import StandardScaler  # HAndling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
```

```python
num_pipelines=Pipeline(
    steps=[
        ("impute",SimpleImputer()),
        ("scaler",StandardScaler())
    ]
)
```

```python
preprocessor=ColumnTransformer(
    [
        ("num_pipelines",num_pipelines,numerical_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=343)
```

```python
preprocessor.fit_transform(x_train)
```

```python
preprocessor.transform(x_test)
```

```python
preprocessor.get_feature_names_out()
```

```python
x_train=pd.DataFrame(preprocessor.transform(x_train),columns=preprocessor.get_feature_names_out())
x_test=pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
x_train
```

```python
x_test
```

```python
""""linear regression
lasso
ridge
Elastic net"""
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import numpy as np
def evaluate_model(true_value,pred_value):
    mae=mean_absolute_error(true_value,pred_value)
    mse=mean_squared_error(true_value,pred_value)
    rmse=np.sqrt(mse)
    rscore=r2_score(true_value,pred_value)
    return mae,rmse,rscore
```

```python
models={
    "LinearRegression":LinearRegression(),
    "Lasso":Lasso(),
    "Ridge":Ridge(),
    "ElasticNet":ElasticNet()
}
```

```python
model_list=[]
r2_list=[]
```

```python
for i in range(len(list(models))):
    models=list(models.values())[i]
    models.fit(x_train,y_train)
    
    #make predictions
    y_pred=models.predict(x_test)
    
    #this is a validation(test) score
    mae,rmse,r2_score=evaluate_model(y_test,y_pred)
    
    print(list(models.keys())[i])
    model_list.append(models.keys()[i])

    print("Model Training Performance")
    print("RMSE:",rmse)
    print("MASE:",mae)
    print("R2 score:",r2_score)
    
    r2_list.append(r2_score)
    
    print("="*35)
    print('\n')
        
```
